```python
import os
import shutil

# Paths
src_root = "/Users/annisaafitrin/Downloads/split_dataset_june-3"
dst_root = "split_dataset"

# Files to copy
files_to_copy = [
    "audio_embedding_gru.npy",
    "labram_base_embeddings.npy",
    "cbramod_embeddings_mumtaz.npy",
    "cbramod_embeddings.npy"

]

# Loop over subject folders
for subject_id in os.listdir(src_root):
    src_folder = os.path.join(src_root, subject_id)
    dst_folder = os.path.join(dst_root, subject_id)

    # Skip if not a folder
    if not os.path.isdir(src_folder):
        continue

    # Ensure destination folder exists
    os.makedirs(dst_folder, exist_ok=True)

    # Copy each file if it exists
    for filename in files_to_copy:
        src_file = os.path.join(src_folder, filename)
        dst_file = os.path.join(dst_folder, filename)

        if os.path.exists(src_file):
            shutil.copy(src_file, dst_file)
            print(f"Copied {src_file} → {dst_file}")
        else:
            print(f"⚠ File not found: {src_file}")

print("✅ Done copying embeddings.")
```
